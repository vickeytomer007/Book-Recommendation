```python
import numpy as np
import pandas as pd
```

```python
df_Ratings=pd.read_csv('BX-Book-Ratings.csv',sep=';',error_bad_lines=False,encoding='latin-1')
df_Books=pd.read_csv('BX-Books.csv',sep=';',error_bad_lines=False,encoding='latin-1')
df_User=pd.read_csv('BX-Users.csv',sep=';',error_bad_lines=False,encoding='latin-1')
```

```python
df_User.head(2)
```

```python
df_Books.head(2)
```

```python
#df1.merge(df2,on='name').merge(df3,on='name')
df_merge=df_Books.merge(df_Ratings,on='ISBN').merge(df_User,on='User-ID')
```

```python
df_merge.head(5)
```

```python
#Remove extra columns that are of no use for recommendation.
df_merge.drop(['Image-URL-S','Image-URL-M','Image-URL-L'],axis=1,inplace=True)
```

```python
df_merge.head(5)
```

```python
#Grouped data on the basis of title ratings.
df_merge.groupby('Book-Title')['Book-Rating'].mean().sort_values(ascending=False).head()
```

```python
ratings=pd.DataFrame(df_merge.groupby('Book-Title')['Book-Rating'].mean())
    
```

```python
ratings['num of ratings']=df_merge.groupby('Book-Title')['Book-Rating'].count()
```

```python
ratings.head()
```

```python
r=ratings.reset_index()
r.head()
```

```python
r[r['Book-Title']=='Clara Callan']
```

```python
df_pivot=df_merge.head(100000).pivot_table(index='User-ID',columns='Book-Title',values='Book-Rating')
#df_merge.head()
```

```python
df_pivot=df_pivot.fillna(0)
df_pivot.head()
```

```python
df_book=df_pivot['13 99 Euros']
```

```python
df_book
similar_to_Euros=df_pivot.corrwith(df_book)
```

```python
similar_to_Euros.head()
```

```python
corr_Euros=pd.DataFrame(similar_to_Euros,columns=['Correlation'])
```

```python
corr_Euros.head()
```

```python
corr_Euros.sort_values('Correlation',ascending=False).head()
```

```python
corr_Euros=corr_Euros.join(ratings['num of ratings'])
```
